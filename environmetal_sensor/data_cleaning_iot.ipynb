```python
metadata={
  "version":"5.3.0",
  "database":"public",
  "container":"device1",
  "containerType":"TIME_SERIES",
  "containerFileType":"csv",
  "containerFile": "public.device1.csv",
  "columnSet":[],
  "rowKeySet":["ts"],
  "timeSeriesProperties":{
    "compresssionMethod":"NO",
    "compressionWindowSize":-1,
    "compressionWindowSizeUnit":"null",
    "expirationDivisionCount":-1,
    "rowExpirationElapsedTime":-1,
    "rowExpirationTimeUnit":"null"
  },
  "compressionInfoSet":[]
}
```

```python
def generate_schema(dataframe) -> list:
    schema = []
    for column_name, dtype in dataframe.dtypes.items():
      if column_name == 'ts':
          field_type='timestamp'
      elif dtype == 'int64':
          field_type = 'INTEGER'
      elif dtype == 'float64':
          field_type = 'double'
      elif dtype == 'bool':
          field_type = 'bool'
      elif dtype == 'datetime64':
          field_type = 'timestamp'
      else:
          field_type = 'STRING'
      if dtype=='object':
        is_nullable=True
      else:
        is_nullable=False
      schema.append({
          'columnName': column_name,
          'type': field_type,
          'notNull': is_nullable
      })
    return schema
```

```python
generate_schema(device1)
```

```python
metadata['columnSet'] = generate_schema(device1)
```
